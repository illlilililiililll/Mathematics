```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

x = np.arange(-10, 10, 0.1)
f_x = f(x)

# 미분
x_new = 10
derivative = []
y = []
learning_rate = 0.1
for i in range(100):
    old_value = x_new
    derivative.append(old_value-learning_rate*2*old_value)
    x_new = old_value-learning_rate*2*old_value
    y.append(x_new**2)

plt.plot(x,f_x)
plt.scatter(derivative,y)
plt.plot(derivative,y, color='red', linestyle=':')
plt.show()
```

```python
# learning rate 값 변화
x_new = 10
derivative = []
y = []
learning_rate = 0.9
for i in range(20):
    old_value = x_new
    derivative.append(old_value-learning_rate*2*old_value)
    x_new = old_value-learning_rate*2*old_value
    y.append(x_new**2)

plt.plot(x,f_x)
plt.scatter(derivative,y)
plt.plot(derivative,y, color='red', linestyle=':', alpha=0.5)
plt.show()
```

$$f(x) = x^3 - 6x^2 + 11x  - 6$$

```python
def tri(x):
    return x**3 - 6*x**2 + 11*x  - 6

def derivative_f(x):
    return 3*x**2 - 12*x + 11
```

```python
x = np.arange(-0.3,4,0.01)
f_x = tri(x)

plt.plot(x, f_x)
plt.plot(x, [0 for i in range(len(x))], color='black')
plt.show()
```

```python
x_new = 5
learning_rate = 0.016
iterations = 100

derivative = []
y = []


for i in range(iterations):
    old_value = x_new
    x_new = old_value - learning_rate * derivative_f(old_value)
    derivative.append(x_new)
    y.append(tri(x_new))

x = np.linspace(-0.3, 5, 100)
f_x = tri(x)


plt.plot(x, f_x)
plt.scatter(derivative, y, color='red')
plt.plot(derivative, y, color='red', linestyle=':', alpha=0.5)
plt.show()
```

$$y = x sin(x^2) + 1$$

```python
def sin_function(x):
    return x*np.sin(x**2)+1

def derivative_f(x):
    return np.sin(x**2) + 2*(x**2) * np.cos(x**2)
```

```python
x = np.arange(-3,3,0.001)
f_x = sin_function(x)
plt.plot(x,f_x)
plt.show()
```

```python
x_new = 1
derivative = []
y = []
learng_rate= 0.01
for i in range(10000):
    old_value = x_new
    x_new = old_value - learng_rate * derivative_f(old_value)
    derivative.append(x_new)
    y.append(sin_function(x_new))

plt.plot(x, f_x)
plt.scatter(derivative, y, color='red')
plt.plot(derivative, y, color='red', linestyle=':', alpha=0.5)
plt.show()
```
